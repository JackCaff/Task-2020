```python
import numpy as np
import collections

def dicerolls(k,n): # Declaring the function dicerolls(): 
    
    D1 = k * 1 # Assign the dice values range from 1
    D2 = k * 6 # Assign the dice values range to 6

    x = np.random.default_rng()
    
    y = (x.integers(D1,D2,size=n,endpoint=True)) # Range D1 & D2 inclusive as endpoint=True
    
    new_y = dict((collections.Counter(y))) # counts the number of times a value is rolled
  
    return new_y # return the output of the function dicerolls(k,n)
```

```python
# [4] & [5] program reference for calling the code fuctions
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy.stats import norm


def coinflips(n): # Create a function Coin flips that allows a user select the number of times the coin is flipped
     
    rng = np.random.default_rng()
    
    x = (rng.binomial(n,p=0.5,size = 1000)) #generate random values to simulate coin flips
    
    heads = rng.binomial(n,0.5,1) #return the number of times head is the result.
    
    print ("Number of times the result is Heads", heads)
    
    # Formatting graph returned [6] & [7]. 
    sns.set_style("darkgrid")
    ax =sns.distplot(x,bins =10, kde=True,fit = norm, color ='r',hist=True) #Formatting the plot. 
    plt.xlabel("Count")
    plt.ylabel('Proability')
    plt.title("Coin Flip Simulation")
    return ax #return the required plot
```
